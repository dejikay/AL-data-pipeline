```python
import pandas as pd
import glob
import os
```

```python
# Define read csv function to read all csv files as objects
def read_csv_mod(file):
    return pd.read_csv(file, dtype=object)
```

```python
# List all files in Versions folder
Versions_file_List = os.listdir(os.getcwd() + "\\versions")
Versions_file_List[0:10]
```

```python
# merging the versions files
Path_Versions_file_List = [os.getcwd() + "\\versions\\" + i for i in Versions_file_List]
df = pd.concat(map(read_csv_mod, Path_Versions_file_List), ignore_index=True)

# Save veriosns files as parquet
df.to_parquet('combined_verions_data.parquet')

print(df)
```

```python
# List all files in timesheets folder
Timesheets_file_List = os.listdir(os.getcwd() + "\\timesheets")
Timesheets_file_List[0:10]
```

```python
# merging the timesheets files
Path_timesheets_file_List = [os.getcwd() + "\\timesheets\\" + i for i in Timesheets_file_List]
df = pd.concat(map(read_csv_mod, Path_timesheets_file_List), ignore_index=True)

# Save veriosns files as parquet
df.to_parquet('combined_timesheets_data.parquet')

print(df)
```

```python
# Combine additional data to Timesheet_Data
Timesheet_Data_Mod = Timesheet_Data.merge(crew_data, on = 'username', how = 'left') # 1. add employee type
Timesheet_Data_Mod = Timesheet_Data_Mod.merge(dept_dept_data, on = 'dept', how = 'left') # 2. add department*
Timesheet_Data_Mod = Timesheet_Data_Mod.merge(depts_fullname_data, on = 'department', how = 'left') # 3. add department full name

# Change Columns to Correct data type
# Timesheet_Data_Mod['entries'] = pd.to_numeric(Timesheet_Data_Mod['entries'])
Timesheet_Data_Mod = Timesheet_Data_Mod.convert_dtypes() # Convert all columns to string
Timesheet_Data_Mod[['entries', 'days_logged', 'days_signed_off']] = Timesheet_Data_Mod[['entries', 'days_logged', 'days_signed_off']].apply(pd.to_numeric) # Convert some columns to numeric
Timesheet_Data_Mod['date'] = pd.to_datetime(Timesheet_Data_Mod['date'], format='%Y-%m-%d')
Timesheet_Data_Mod['week'] = pd.to_datetime(Timesheet_Data_Mod['week'], format='%Y-%m-%d')
Timesheet_Data_Mod['month'] = pd.to_datetime(Timesheet_Data_Mod['month'], format='%Y-%m-%d')

Timesheet_Data_Mod.to_parquet('combined_Timesheet_Data_Mod.parquet')
display(Timesheet_Data_Mod)
```

```python
# Combine additional data to Versions_Data
Versions_Data_Mod = Versions_Data.merge(rtypes_data, left_on = 'reviewtype', right_on = 'rtype', how = 'left') # 1. add dept and type based on reviewtype
Versions_Data_Mod = Versions_Data_Mod.merge(dept_dept_data, on = 'dept', how = 'left') # 2. add department
Versions_Data_Mod = Versions_Data_Mod.merge(depts_fullname_data, on = 'department', how = 'left') # 3. add department full name
Versions_Data_Mod = Versions_Data_Mod.merge(crew_data, on = 'username', how = 'left') # 4. add employee type

# Change Columns to Correct data type
Versions_Data_Mod = Versions_Data_Mod.convert_dtypes() # Convert all columns to string
Versions_Data_Mod[['count', 'bytes']] = Versions_Data_Mod[['count', 'bytes']].apply(pd.to_numeric) # Convert some columns to numeric
Versions_Data_Mod['week'] = pd.to_datetime(Versions_Data_Mod['week'], format='%Y-%m-%d')
Versions_Data_Mod['month'] = pd.to_datetime(Versions_Data_Mod['month'], format='%Y-%m-%d')

# Versions_Data_Mod.to_parquet('combined_Versions_Data_Mod.parquet')
display(Versions_Data_Mod)
```
